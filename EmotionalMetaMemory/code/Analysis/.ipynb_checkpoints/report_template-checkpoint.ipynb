```python

import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pyExG.detection import oxi_peaks
from pyExG.utils import heart_rate
from trials2counts import trials2counts
from fit_meta_d_MLE import fit_meta_d_MLE
sns.set_context('talk')
path = os.getcwd()
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

# d' & Meta d'

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace=.5, wspace=.001)
axs = axs.ravel()
i = 0
for ar in ['Low', 'High']:
    for val in ['Low', 'High']:
        df2 = df.copy()[(df.Arousal == ar) & (df.Valence == val)]

        stimID = df2.Expected == 'right'
        response = df2.KeyPressed == 'right'
        rating = df2.Confidence
        nRatings = 7

        nR_S1, nR_S2 = trials2counts(stimID, response, rating, nRatings,1)
        fit = fit_meta_d_MLE(nR_S1,nR_S2);

        axs[i].plot(nR_S1, 'ro-')
        axs[i].plot(nR_S2, 'go-')
        axs[i].set_title('Arousal ' + ar + ', Valence ' + val + ' Meta d = ' + str(round(fit['meta_da'], 2)))
        i += 1
plt.tight_layout()
```

Find all the files ending with '.npy' (oximeter recording), and store the name in a list.

```python
oxi_files = []
for file in os.listdir(path + '/Data/' + subject):
    if file.endswith(".npy"):
        oxi_files.append(file)
```

```python
assert(len(oxi_files) == 12)
```

**Plot the first seconds of PPG recording**

```python
sns.set_context('paper')
fig, axs = plt.subplots(3, 4, figsize=(15, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace=.5, wspace=.001)
axs = axs.ravel()
for i in range(len(oxi_files)):
    oxi = np.load(path + '/Data/' + subject + '/' + oxi_files[i], allow_pickle=True)
    oxi = moving_average(oxi[0], 20)
    axs[i].plot(oxi[:500])
    axs[i].set_title('Block' + str(i+1) + ', lenght=' + str(round(len(oxi)/75)) + ' s.')
plt.tight_layout()
```

**Heart rate time series**

```python
sns.set_context('paper')
fig, axs = plt.subplots(3, 4, figsize=(15, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)
axs = axs.ravel()
for i in range(len(oxi_files)):
    oxi = np.load(path + '/Data/' + subject + '/' + oxi_files[i], allow_pickle=True)
    oxi = moving_average(oxi[0], 20)
    peaks = oxi_peaks(oxi)
    hr, time = heart_rate(peaks, sfreq=75, unit='bpm', method='staircase')
    axs[i].plot(hr)
    axs[i].set_title('Block' + str(i+1))
    axs[i].set_ylabel('BPM')
plt.tight_layout()
```
