```python
from scipy.stats import norm
import math
Z = norm.ppf
 
def SDT(hits, misses, fas, crs):
    """Returns d-prime measure given hits, misses,
    false alarms, and correct rejections
    
    Parameters
    ----------
    hits : float
        Hits.
    misses :  float
        Misses.
    fas : float
        False alarms.
    crs : float
        Correct rejections.
    Returns
    -------
    d : d' value
    """
    # Floors an ceilings are replaced by half hits and half FA's
    half_hit = 0.5 * (hits + misses)
    half_fa = 0.5 * (fas + crs)
 
    # Calculate hit_rate and avoid d' infinity
    hit_rate = hits / (hits + misses)
    if hit_rate == 1: 
        hit_rate = 1 - half_hit
    if hit_rate == 0: 
        hit_rate = half_hit
 
    # Calculate false alarm rate and avoid d' infinity
    fa_rate = fas / (fas + crs)
    if fa_rate == 1: 
        fa_rate = 1 - half_fa
    if fa_rate == 0: 
        fa_rate = half_fa
 
    # Return d'
    d = Z(hit_rate) - Z(fa_rate)
    return d
```

```python
df = pd.read_csv('total.txt', sep=',')
```

```python
df = df.dropna()
df = df[df.RT>0]
```

```python
sns.violinplot(data=df, x='Arousal', y='RT', hue='Valence', dodge=True)
```

```python
dprime_df = pd.DataFrame([])
for nSub in df.Subject.unique():
    for ar in ['Low', 'High']:
        for val in ['Low', 'High']:

            this_df = df.copy()[(df.Arousal == ar) & (df.Valence == val) & (df.Subject == nSub)]

            # noise + signal + response: HIT
            hits = sum((this_df.Expected == 'right') & (this_df.KeyPressed == 'right')) / len((this_df.Expected == 'right') & (this_df.KeyPressed == 'right'))

            # noise + signal + no response: MISS
            misses = sum((this_df.Expected == 'right') & (this_df.KeyPressed == 'left')) / len((this_df.Expected == 'right') & (this_df.KeyPressed == 'left'))

            # noise + response: FALSE ALARM (FA)
            fas = sum((this_df.Expected == 'left') & (this_df.KeyPressed == 'right')) / len((this_df.Expected == 'left') & (this_df.KeyPressed == 'right'))

            # noise + no response: CORRECT REJECTION (CR)
            crs = sum((this_df.Expected == 'left') & (this_df.KeyPressed == 'left')) / len((this_df.Expected == 'left') & (this_df.KeyPressed == 'left'))

            assert(hits + misses + fas + crs)

            d = SDT(hits, misses, fas, crs)
            if d == float("inf"):
                out = d

            stimID =    this_df.Expected == 'right'
            response =  this_df.KeyPressed == 'right'
            rating =    this_df.Confidence
            nRatings = 7

            nR_S1, nR_S2 = trials2counts(stimID, response, rating, nRatings,1)
            fit = fit_meta_d_MLE(nR_S1,nR_S2);
            dprime_df = dprime_df.append(pd.DataFrame({'Subject': [nSub],
                                                       'Arousal': [ar],
                                                       'Valence': [val],
                                                       'dprime': [d],
                                                       'Metadprime': [fit['meta_da']]}))
```
